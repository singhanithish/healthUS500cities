```python
df_smoking_geo = df2016cities.loc[df2016cities['Short_Question_Text']=='Current Smoking',['CityName','StateDesc','Data_Value','Geolocation']]
```

```python
df_smoking_group = df_smoking_geo.groupby(by='CityName',as_index='False').agg({'StateDesc' :'first', 'Data_Value' : 'mean','Geolocation':'first'})
```

```python
df_smoking_group.head(10)
```

```python
df_smoking_group['longitude']= df_smoking_group['Geolocation'].apply(lambda x : eval(x)[0])
```

```python
df_smoking_group['latitude']= df_smoking_group['Geolocation'].apply(lambda x : eval(x)[1])
```

```python
df_smoking_group.head(10)
```

```python
gdf = gpd.GeoDataFrame(
    df_smoking_group, geometry=gpd.points_from_xy(df_smoking_group.latitude, df_smoking_group.longitude))
```
